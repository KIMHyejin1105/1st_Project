```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv("./dataset/must_use_final.csv")

# 2. TIMESTAMP ÌòïÏãù Î≥ÄÌôò Î∞è Ï†ïÎ†¨
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
df = df.sort_values(by=['VSL_ID', 'PORT_NAME', 'TIMESTAMP']).reset_index(drop=True)

# 3. ÏÑ†Î∞ï + Ìï≠Íµ¨ Ï°∞Ìï©ÏúºÎ°ú Í≥†Ïú† ÏãùÎ≥ÑÏûê ÏÉùÏÑ±
df['VSL_PORT_KEY'] = df['VSL_ID'] + "_" + df['PORT_NAME']

# 4. Í∞Å Í∑∏Î£πÎ≥ÑÎ°ú 84Î≤àÏß∏(=14ÏãúÍ∞Ñ ÏãúÏ†ê) Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
rows_14h = []
for key, group in df.groupby('VSL_PORT_KEY'):
    if len(group) >= 84:
        row_14h = group.iloc[83]
        rows_14h.append(row_14h)

df_14h = pd.DataFrame(rows_14h).reset_index(drop=True)

# 5. Íµ∞ÏßëÎ≥Ñ Ìï≠Íµ¨Î≥Ñ Í∞úÏàò Î∞è Íµ∞Ïßë Ï¥ùÌï© Í≥ÑÏÇ∞
grouped_detail = df_14h.groupby(['CLUSTER_1', 'PORT_NAME']).size().reset_index(name='count_14h')
grouped_total = df_14h.groupby('CLUSTER_1').size().reset_index(name='cluster_total')
merged = pd.merge(grouped_detail, grouped_total, on='CLUSTER_1')

# 6. Íµ∞ÏßëÎ™ÖÏóê Ï¥ùÌï© Î∂ôÏù¥Í∏∞
merged['CLUSTER_1_LABEL'] = merged.apply(
    lambda row: f"CLUSTER_1 = {row['CLUSTER_1']} (Ï¥ù {row['cluster_total']}Í∞ú)", axis=1
)

# 7. Î≥¥Í∏∞ Ï¢ãÍ≤å Ï∂úÎ†•
grouped_print_data = merged.groupby('CLUSTER_1_LABEL')

for cluster_label, group in grouped_print_data:
    print(f"üëâ {cluster_label}")
    print(group[['PORT_NAME', 'count_14h']].to_string(index=False))
    print()
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ (Íµ∞Ïßë 1Îßå Ìè¨Ìï®Îêú ÌååÏùº)
df = pd.read_csv("./dataset/cluster1_6.csv")

# 2. TIMESTAMP Ï†ïÎ†¨
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
df = df.sort_values(by=['VSL_ID', 'PORT_NAME', 'TIMESTAMP']).reset_index(drop=True)

# 3. ÏÑ†Î∞ï+ÎèÑÏ∞©Ìï≠Íµ¨ Ï°∞Ìï©Î≥ÑÎ°ú Í∑∏Î£πÌïë ÌõÑ 84Î≤àÏß∏(=14ÏãúÍ∞Ñ) ÏãúÏ†ê Ï∂îÏ∂ú
rows_14h = []
for (vsl_id, port), group in df.groupby(['VSL_ID', 'PORT_NAME']):
    if len(group) >= 84:
        row_14h = group.iloc[83]
        rows_14h.append(row_14h)

df_14h = pd.DataFrame(rows_14h).reset_index(drop=True)

# 4. Ìï≠Íµ¨Î≥Ñ Í∞úÏàò Í≥ÑÏÇ∞
port_counts = df_14h['PORT_NAME'].value_counts().reset_index()
port_counts.columns = ['PORT_NAME', 'count_14h']

# 5. Ï¥ùÌï© Í≥ÑÏÇ∞
total = port_counts['count_14h'].sum()

# 6. ÏòàÏÅòÍ≤å Ï∂úÎ†•
print(f"üì¶ CLUSTER_1 = 1 (Ï¥ù {total}Í∞ú)")
print(port_counts.to_string(index=False))
```
